```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Number of vehicles belonging to each brand

```python
# Creating a Plot that shows the number of vehicles belonging to each brand
sns.set_style("whitegrid")
g = sns.catplot(y="brand", data=df, kind="count", palette="Reds_r", height=7, aspect=1.5)
g.ax.set_title("Vehicles By Brand",fontdict={'size':18})
g.ax.xaxis.set_label_text("Number of Vehicles",fontdict= {'size':16})
g.ax.yaxis.set_label_text("Brand",fontdict= {'size':16})
plt.show()
```

```python
# Saving the plot
g.savefig(("plots/Analise2/brand-vehicleCount.png"))
```

## Average vehicle price based on vehicle type as well as gearbox type

```python
# Plotting the Average Price of Vehicles Based on Vehicle Type as well as Gearbox Type
fig, ax = plt.subplots(figsize=(8,5))
colors = ["#00e600", "#ff8c1a","#a180cc"]
sns.barplot(x="vehicleType", y="price",hue="gearbox", palette=colors, data=df)
ax.set_title("Average vehicle price by vehicle type and gearbox type",fontdict= {'size':12})
ax.xaxis.set_label_text("Vehicle Type",fontdict= {'size':12})
ax.yaxis.set_label_text("Average price",fontdict= {'size':12})
plt.show()
```

```python
# Saving the plot
fig.savefig("plots/Analise2/vehicletype-gearbox-price.png")
```
