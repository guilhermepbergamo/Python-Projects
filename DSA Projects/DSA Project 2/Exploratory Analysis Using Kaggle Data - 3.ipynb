```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Average vehicle price by fuel type and gearbox type

```python
# Creating a Barplot with Average Vehicle Price by Fuel Type and Gearbox Type
fig, ax = plt.subplots(figsize=(8,5))
colors = ["#00e600", "#ff8c1a","#a180cc"]
sns.barplot(x="fuelType", y="price",hue="gearbox", palette="husl",data=df)
ax.set_title("Average vehicle price by fuel type and gearbox type",fontdict= {'size':12})
ax.xaxis.set_label_text("Fuel Type",fontdict= {'size':14})
ax.yaxis.set_label_text("Average Price",fontdict= {'size':14})
plt.show()
```

```python
# Saving the plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price.png")
```

## Average vehicle power by vehicle type and gearbox type

```python
# Creating a Barplot with the Average Vehicle Power by Vehicle Type and Gearbox Type
colors = ["windows blue", "amber", "greyish", "faded green", "dusty purple"]
fig, ax = plt.subplots(figsize=(8,5))
sns.set_palette(sns.xkcd_palette(colors))
sns.barplot(x="vehicleType", y="powerPS",hue="gearbox",data=df)
ax.set_title("Average vehicle power by vehicle type and gearbox type",fontdict= {'size':12})
ax.xaxis.set_label_text("Vehicle Type",fontdict= {'size':14})
ax.yaxis.set_label_text("Average Power",fontdict= {'size':14})
plt.show()
```

```python
# Saving the plot
fig.savefig("plots/Analise3/vehicletype-fueltype-power.png")
```
